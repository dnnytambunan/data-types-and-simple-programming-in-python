Daftar Paket Menu Rumah Makan Numphy Sejahtera
1. Ayam Goreng Numphy + Nasi + Es teh Manis
2. Ayam Bakar Numphy + Nasi + Es Teh Manis
3. Bebek Goreng + Nasi + Es Teh Manis
4. Bebek Bakar + Nasi + Es Teh Manis
5. Ayam Kecap Numphy + Nasi + Es Jeruk Manis
6. Ayam rica-rica + Nasi + + Es Teh Manis

```python
#coding pemilihan menu pelanggan
menu = int(input("Pilih Menu Anda = "))

if menu == 1:
  print("Ayam Goreng Numphy + Nasi + Es teh Manis \nRp22000")
elif menu == 2:
  print("Ayam Bakar Numphy + Nasi + Es Teh Manis \nRp24000")
elif menu == 3:
  print("Bebek Goreng + Nasi + Es Teh Manis \nRp25000")
elif menu == 4:
  print("Bebek Bakar + Nasi + Es Teh Manis \nRp25000")
elif menu == 5:
  print("Ayam Kecap Numphy + Nasi + Es Jeruk Manis \nRp24000")
elif menu == 6:
  print("Ayam rica-rica + Nasi +  Es Teh Manis \nRp24000")
else:
  print("Maaf, Belum ada menu paket lainnya")
```
